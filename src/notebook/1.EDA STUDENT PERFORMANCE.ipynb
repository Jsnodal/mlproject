#### Numerical and categrical

```python
# This code print down   the numerical  and categorical features
 

numerical_features = [feature for feature in data.columns if data[feature].dtypes != 'object']
categorical_features = [feature for feature in data.columns if data[feature].dtypes =='object']
print("Numerical features: {}".format(numerical_features))
print("Categorical features: {}".format(categorical_features))

 
```

#### Adding column for Total score and Average

```python
#feature engieering, engineering ur total scores and average scores

data["total score"] = data["math score"]+data["writing score"]+data["reading score"]
data["average"] = data["total score"]/3
data.head()
```

#### Student and their total marks

```python
reading_full = data[data["reading score"] == 100]["average"].count()
writing_full = data[data["writing score"]==100]["average"].count()
math_full = data[data["math score"]==100]["average"].count()

print(f"Number of student with full makrs in Math :{math_full}")
print(f"Number of student with full marks in Writing : {writing_full}")
print(f"Number of student with full marks in reading : {reading_full}")
```

```python
reading_less_20 = data[data["reading score"]<=20]["average"].count()
writing_less_20 = data[data["writing score"]<=20]["average"].count()
math_less_20 = data[data["math score"]<=20]["average"].count()

print(f"Number of student with below 20 makrs in Math :{math_less_20}")
print(f"Number of student with below 20 marks in Writing : {writing_less_20}")
print(f"Number of student with below 20 marks in reading : {reading_less_20}")

```

#### Students Scoring Below 20 Marks:

Math: 4 students scored below 20.
Writing: 3 students scored below 20.
Reading: 1 student scored below 20.
Key Observations:

Math has the highest number of students (4) scoring below 20, indicating that it may be a challenging subject for many.
Writing follows closely, with 3 students scoring below 20.
Reading has the least number of students (1) with scores below 20, suggesting that students are performing better in this area compared to Math and Writing.
Implications:

There may be a need for additional support and resources in Math and Writing to help students improve their scores.
Reading appears to be a strength for most students, which can be leveraged to enhance their performance in other subjects.

# Visualization

```python
# Histogram and kde
fig,axs = plt.subplots(1, 2,figsize = (15,7))
plt.subplot(121)
sns.histplot(data,x ="average",bins = 30, kde = True, color= "g")
plt.subplot(122)
sns.histplot(data,x = "average",kde = True,hue = "gender")
plt.show()
```

## Maximum score of allstudents

```python
plt.figure(figsize=(18,18))
plt.subplot(1,4,1)
plt.title("MATH SCORE")
sns.violinplot(data ,y = "math score" ,color = "red",linewidth = 3)
plt.subplot(1,4,2)
plt.title("READING SCORES")
sns.violinplot(data, y= "reading score",color = "green",linewidth = 3)
plt.subplot(1,4,3)
plt.title("WRITING SCORES")
sns.violinplot(data,y = "writing score",color = "blue",linewidth=3)
plt.show()
```

#### Is gender has any impact of student perfomance

```python
# Select numerical features (float64) and group by gender, then calculate the mean
gender_group = data.select_dtypes(include=['float64','int64']).groupby(data["gender"]).mean()
print(gender_group)

```

```python
plt.figure(figsize=(10, 6))
gender_group.plot(kind='bar', color=['skyblue',"red" ,"violet",'lightgreen', 'salmon'], edgecolor='black')
 
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.grid(axis='y')

 
plt.tight_layout()
plt.show()
```

#### On an avarge female has better score or higher score than males

## Is our data skewed?

```python
skewness_results = {}
for column in data.select_dtypes(include=['int64', 'float64']).columns:
    skewness_results[column] = data[column].skew()

# Display skewness results
print("Skewness of numerical columns:")
for column, skew in skewness_results.items():
    print(f"{column}: {skew}")
```
